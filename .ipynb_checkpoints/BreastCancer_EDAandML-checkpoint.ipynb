<h2 align=center>Tumor Diagnosis: Exploratory Data Analysis and Machine Learning</h2>

### About the Dataset:

The [Breast Cancer Diagnostic data](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29) is available on the UCI Machine Learning Repository. This database is also available through the [UW CS ftp server](http://ftp.cs.wisc.edu/math-prog/cpo-dataset/machine-learn/cancer/WDBC/).

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. 

**Attribute Information**:

- ID number
- Diagnosis (M = malignant, B = benign)

Ten real-valued features are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter) 
2. texture (standard deviation of gray-scale values) 
3. perimeter 
4. area 
5. smoothness (local variation in radius lengths) 
6. compactness (perimeter^2 / area - 1.0) 
7. concavity (severity of concave portions of the contour) 
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant

###  Machine learning models

First lets train a simple support vector classifer (SVC) to see what effect does normalization have on the model.

```python
y.replace({'M':1,'B':0}, inplace=True) # making the target labels numerical for efficient analysis
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import *

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
                                                   
SVCmodel_un = SVC(C=1).fit(X_train, y_train)  #on the default rbf kernel
print('Breast cancer with unnormalized features')
print('Accuracy of SVC on training set: {:.2f}'
     .format(SVCmodel_un.score(X_train, y_train)))
print('Accuracy of SVC on test set: {:.2f}'
     .format(SVCmodel_un.score(X_test, y_test)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(data_n_2, y, random_state = 0)
                                                   
SVCmodel_n = SVC(C=1).fit(X_train, y_train)  #on the default rbf kernel
print('Breast cancer with unnormalized features')
print('Accuracy of SVC on training set: {:.2f}'
     .format(SVCmodel_n.score(X_train, y_train)))
print('Accuracy of SVC on test set: {:.2f}'
     .format(SVCmodel_n.score(X_test, y_test)))
print('Recall score of SVC on test set: {:.2f}'
     .format(recall_score(y_test,SVCmodel_n.predict(X_test))))
print('Roc score of SVC on test set: {:.2f}'
     .format(roc_auc_score(y_test,SVCmodel_n.predict(X_test))))
```

Significant improvement on normalizing the features as expected.

```python
x1 = data_n_2[uncorrelated_vars]

X_train, X_test, y_train, y_test = train_test_split(x1, y, random_state = 0)
                                                   
SVCmodel_n = SVC(C=1).fit(X_train, y_train)  #on the default rbf kernel
print('Breast cancer with normalized features')
print('Accuracy of SVC on training set: {:.2f}'
     .format(SVCmodel_n.score(X_train, y_train)))
print('Accuracy of SVC on test set: {:.2f}'
     .format(SVCmodel_n.score(X_test, y_test)))
print('Recall score of SVC on test set: {:.2f}'
     .format(recall_score(y_test,SVCmodel_n.predict(X_test))))
print('Roc score of SVC on test set: {:.2f}'
     .format(roc_auc_score(y_test,SVCmodel_n.predict(X_test))))
```

No improvement as in accuracy. But clearly the recall and roc scores improve (for a fixed C) owing to the fact that we choose features wisely.

Now lets try out a decision tree model - Gradient boosted decision tree which is similar to a random forest model in that it has several trees but unike the random forest model, the trees here are trained in series. That is, the next tree learns from the previous tree's mistake and corrects itself in succesive iterations. Also, a clear advantage of a decision based tree is the non-reliance on the feature normalization. 

```python
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0) 
                                                   

GBDT = GradientBoostingClassifier(learning_rate = 0.1, max_depth = 3,random_state = 0)
GBDT.fit(X_train, y_train)

print('Breast cancer dataset (learning_rate=0.1, max_depth=3)')
print('Accuracy of GBDT classifier on training set: {:.2f}'
     .format(GBDT.score(X_train, y_train)))
print('Accuracy of GBDT classifier on test set: {:.2f}\n'
     .format(GBDT.score(X_test, y_test)))

GBDT = GradientBoostingClassifier(n_estimators = 100,learning_rate = 0.01, max_depth = 2, random_state = 0)
GBDT.fit(X_train, y_train)

print('Breast cancer dataset (learning_rate=0.01, max_depth=2)')
print('Accuracy of GBDT classifier on training set: {:.2f}'
     .format(GBDT.score(X_train, y_train)))
print('Accuracy of GBDT classifier on test set: {:.2f}'
     .format(GBDT.score(X_test, y_test)))
```

### GridSearch analysis for optmizing our classifier

Now lets use a grid search with both the classifiers considered above to see what hyperparameters works the best. Also for a disease like cancer where a false negative could have pretty serious implications, accuracy is not the best metric to judge a model. Instead, metrics like the roc_score and recall are more valuable. So, we will explore those as well.

```python
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import roc_auc_score

#y.replace({'M':1,'B':0}, inplace=True)
X_train, X_test, y_train, y_test = train_test_split(data_n_2, y, random_state=0)
X_train2, X_test2, y_train2, y_test2 = train_test_split(x1, y, random_state=0)

SVCmodel = SVC().fit(X_train,y_train) #default kernel
SVCmodel2 = SVC().fit(X_train2,y_train2) #default kernel

grid_values = {'C': [0.001, 0.01, 0.05, 0.1, 1, 10, 100], 'gamma':[0.001, 0.01, 0.05, 0.1, 1, 10, 100],
               'class_weight':[None,'balanced', {1:2},{1:3},{1:4},{1:5},{1:10},{1:20},{1:50}]}

for i, metric in enumerate(('accuracy','recall', 'f1','roc_auc')):
    grid_SVC = GridSearchCV(SVCmodel, param_grid=grid_values, cv=3, scoring=metric)
    grid_SVC2 = GridSearchCV(SVCmodel2, param_grid=grid_values, cv=3, scoring=metric)
    grid_SVC.fit(X_train, y_train)
    grid_SVC2.fit(X_train2, y_train2)
    print('Grid best parameter for all features (max. {0}): {1}'
          .format(metric, grid_SVC.best_params_))
    print('Grid best score for all features({0}): {1}'
          .format(metric, grid_SVC.best_score_))
    print('Grid best parameter for chosen features (max. {0}): {1}'
          .format(metric, grid_SVC2.best_params_))
    print('Grid best score for chosen features({0}): {1}'
          .format(metric, grid_SVC2.best_score_))
```

One would probably look to increase the sensitivity/recall in this case as we want to reduce the number of false negatives.

So, in this case the model predicted by the 'recall' metric would be the best one to go with. But if we also want to reduce false positives as well along with a high recall, roc_score would be the one to look at. Also, overfitting here does not really hurt the performance as it fits to some features which aren't there in one of the models. It does not really contradict what we showed earlier - that was for a fixed C of 1.

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0) 

GBDTmodel = GradientBoostingClassifier().fit(X_train,y_train) #default kernel
grid_values = {'learning_rate': [0.001, 0.01, 0.05, 0.1, 1, 10], 'max_depth':[1, 2, 3, 4, 5], 
               'max_features':[5,7,10,15,20,25,30]}

for i, metric in enumerate(('accuracy','recall', 'f1','roc_auc')):
    grid_GBDT = GridSearchCV(GBDTmodel, param_grid=grid_values, cv=3, scoring=metric)
    grid_GBDT.fit(X_train, y_train)
    print('Grid best parameter (max. {0}): {1}'
          .format(metric, grid_GBDT.best_params_))
    print('Grid best score ({0}): {1}'
          .format(metric, grid_GBDT.best_score_))
```

```python
from sklearn.metrics import plot_roc_curve

X_train, X_test, y_train, y_test = train_test_split(data_n_2, y, random_state=0)
X_train2, X_test2, y_train2, y_test2 = train_test_split(x, y, random_state = 0) 
X_train3, X_test3, y_train3, y_test3 = train_test_split(x1, y, random_state = 0) 

final_model1 = SVC(C = 10, gamma = 0.01).fit(X_train,y_train)
final_model1_custom = SVC(C = 10, gamma = 0.01).fit(X_train3,y_train3)
final_model2 = GradientBoostingClassifier(learning_rate = 1, max_depth = 2, max_features = 20).fit(X_train2,y_train2)

svc_disp = plot_roc_curve(final_model1, X_test, y_test)
svc_disp2 = plot_roc_curve(final_model2, X_test2, y_test2,ax=svc_disp.ax_)
svc_disp3 = plot_roc_curve(final_model1_custom, X_test3, y_test3, name = 'SVC_custom', ax=svc_disp.ax_)

plt.show()
```
